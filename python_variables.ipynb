Variable Names

Bir variable’ın kısa bir adı (x ve y gibi) veya daha açıklayıcı bir adı (age, carname, toplam_hacim) olabilir.

Python variable kuralları 
1- Bir variable adı bir harfle veya alt çizgi karakteriyle başlamalıdır.

2- Bir variable adı sayı ile başlayamaz.

3- Bir variable adı yalnızca alfa-sayısal karakterler ve alt çizgiler (Az, 0-9 ve _ ) içerebilir.

4- Variable adları büyük/küçük harfe duyarlıdır (age, Age ve AGE üç farklı değişkendir).

5- Bir variable adı Python anahtar sözcüklerinden hiçbiri olamaz.

Multi Words Variable Names

Camel Case
İlk kelime hariç her kelime büyük harfle başlar.

Pascal Case
Her kelime büyük harfle başlar. Daha yaygın olarak “UpperCamelCase” olarak da bilinir.

Snake Case
Her kelime bir alt çizgi karakteriyle ayrılır.
