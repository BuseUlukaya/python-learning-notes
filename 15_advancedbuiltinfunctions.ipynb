LOOPS

```python
last_list = [10, 20, 30, 40, 50]
while 20 in last_list:
    print("20 in last_list")
    last_list.pop() 
```

20 listenin içinde olduğu sürece döngü çalışır.

RANDOM

```python
from random import randint # Ra#ndom kütüphanesinden randint fonksiyonunu import et demektir.
randint(0,100) # 0 ile 100 arasında rastgele bir sayo getir.
    
```

```python
from random import shuffle #Listeyi rastgele karıştırır.
shuffle(my_list) 
```

```python
my_list = [10, 20, 30, 40,  50, 60, 70]
randint(0, len(my_list)) #0 ile len my_list arasında bir sayı ver dedim.
my_list[randint(0, len(my_list))]
#Eğer randint tam 7 verirse out of index hatası alabilirim. Bunu önlemek için kodu aşağıdaki gibi kurgulamam çok doğru olur.
my_list[randint(0, len(my_list)) -1]
```

ZIP

```python
#Farklı listeleri ilişkilendirmek için zip kullanabilirim.
food_list = ["apple", "banana", "lemon"]
calories_list = [100, 150, 200]
day_list = ["monday", "tuesday", "wednesday" ]
zipped_list = list(zip(food_list, calories_list, day_list))
zipped_list
```
