LIST DATA TYPE

Liste, Python’da: Birden fazla elemanı; Sıralı (ordered), Değiştirilebilir (mutable) şekilde tutan veri tipidir.

Tanımlama:

```python
liste = [1, 2, 3, 4]
```

Listenin Temel Özellikleri

Farklı veri tiplerini birlikte tutabilir.

Sıralıdır → indeks vardır.

Değiştirilebilirdir → eleman eklenebilir, silinebilir, değiştirilebilir.

```python
x = [0, 1, 2, 3, 4, 5]
```

```python
x[1:4]   # [1, 2, 3]
```

```python
x[:3]    # [0, 1, 2]
```

```python
x[3:]    # [3, 4, 5]
```

```python
x[::2]   # [0, 2, 4]
```

Liste mutable olduğu için eleman değiştirilebilir.

```python
x = [10, 20, 30]
x[1] = 99
print(x[1])
```

LİSTEYE ELEMAN EKLEME

append() → sona ekler

```python
x = [1, 2]
x.append(3)   # [1, 2, 3]
print(x)
```

insert(i, değer) → araya ekler

```python
x.insert(1, 99)   # [1, 99, 2, 3]
```

```python
print(x)
```

extend(), bir listeye başka bir listenin tüm elemanlarını tek tek eklemek için kullanılır.

```python
a = [1, 2, 3]
b = [4, 5]

a.extend(b)

print(a)   # [1, 2, 3, 4, 5]
```

```python
fruit = ["Grape", "Orange", "Strawberry", "Pineapple", "Melon", "Watermelon"] + ["cherry", "apple"]

print(fruit)
print(len(fruit))
```

INDEX METODU

index(), bir elemanın listede ilk geçtiği indeks numarasını verir.-> “Bu değer listede kaçıncı sırada?”

```python
x = ["apple", "banana", "cherry"]

i = x.index("banana")
print(i)   # 1
```

COUNT METODU

count(), bir elemanın listede kaç kez geçtiğini sayar.-> “Bu değer listede kaç tane var?”

```python
x = ["apple", "banana", "apple", "cherry", "apple"]

n = x.count("apple")
print(n)   # 3
```

LİSTELERİ SIRALAMA

sort(), bir listenin elemanlarını küçükten büyüğe (artan sırada) yerinde (in-place) sıralar.

Metinler alfabetik sıraya göre sıralanır.

Büyük harf – küçük harf sıralamayı etkiler.

Büyük harfle başlayanlar, küçük harflerden önce gelir.

sort(reverse=True), bir listenin elemanlarını büyükten küçüğe yani azalan sırada sıralamak için kullanılır.Listeyi yerinde değiştirir, yeni liste oluşturmaz.

```python
myNumbers = [70, 19, 34, 23, 7, 34]

myNumbers.sort(reverse=True)

print(myNumbers)
```
