Random Numbers

Pythonda rastgele sayı üreten bir fonksiyon yoktur.

Bu yüzden random modülü üretilmiştir

Random Metodu

```python
import random
print(random.random())
```

Bu şekilde kendi sistem saatimizi kullanarak rastgele sayı üretti.

Seed Metodu

```python
import random
random.seed(87)
print(random.random())
```

Varsayılan değeri sistem saatinden çekmesinin sebebi zaman saniye bazında sürekli değiştiğinden seed değeri sürekli değişiyor. Bu yüzden her seferinde farklı bir rastgele değer üretiyor. Ancak biz seed değerini atayınca bir sabit sayı üzerinden hep aynı rastgele sayıyı üretiyor.

Randint Metodu

```python
import random
print(random.randint(1,10))
```

randint(a, b) metodu, verilen iki sayı arasında(sayıların kendisi de dahil) rastgele bir tam sayı üretir.

Randrange Metodu

randrange(start, stop, step) metodu, belirli bir aralıktan ve belirli bir artışa göre rastgele bir tam sayı üretir.

```python
import random

x = random.randrange(1, 10)
print(x)
```

UYARI: Üst sınır dahil değildir.

```python
x = random.randrange(0, 10, 2)
print(x)
```

Choice Metodu

choice() metodu, bir dizi (list, tuple, string) içinden rastgele tek bir eleman seçer.

```python
import random

x = random.choice([1, 2, 3, 4, 5])
print(x)
```

Choices Metodu

choices() metodu, bir diziden birden fazla rastgele eleman seçer ve sonucu liste olarak döndürür.

```python
import random

x = random.choices([1, 2, 3, 4, 5], k=3)
print(x)
```

UYARI: Aynı eleman tekrar edebilir.

Shuffle Metodu

shuffle() metodu, bir listenin elemanlarını yerinde (in-place) rastgele karıştırır.

```python
import random

numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(numbers)
```

Sample Metodu

sample(population, k) metodu, verilen bir diziden tekrarsız olarak k tane eleman seçer.

```python
import random

x = random.sample([1, 2, 3, 4, 5], 3)
print(x)
```

Uniform Metodu

uniform(a, b) metodu, a ile b arasında rastgele bir ondalıklı sayı (float) üretir.

Teorik olarak uçlar dahil kabul edilir.

Sürekli (float) dağılım olduğu için teorik uçlar tam olarak nadiren çıkar.

```python
import random

print(random.uniform(1, 5))
```

```python
Triangular Metodu
```

triangular(low, high, mode) metodu, low ile high arasında, en olası değeri mode olan üçgensel dağılıma göre rastgele bir float sayı üretir.

```python
import random

x = random.triangular(1, 10, 5)
print(x)
```
