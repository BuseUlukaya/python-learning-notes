BİLGİSAYARLARIN ÇALIŞMA MANTIĞI

Bilgisayar üzerinden elektrik geçen devrelerden oluşan bir sistemdir. Eğer bir devrede elektrik yoksa o devrenin değeri 0 Volt'tur.Elektrik geçtiğinde değeri 5 Volt'tur.

İkili sistemde 0 Volt'u 0 ile 5 Volt'u 1 ile temsil ederiz. Gönderdiğimiz sinyallerle bilgisayarla iletişim kuruyoruz.

ord() fonksiyonu: Bir karakterin ASCII / Unicode sayısal karşılığını verir.

bin() fonksiyonu: Bir onluk (decimal) sayıyı ikili (binary) sisteme çevirir.

```python
print(ord("F"))
print(bin(70))
```

Character Encoding = Karakter kodlama -> Yani Bir harfin, sayının, sembolün bilgisayarda hangi sayıyla temsil edileceğini belirleyen sistem.

Neden encoding’e ihtiyaç var?

Bilgisayar sadece 0 ve 1 bilir.

Ama biz:

Harf yazıyoruz → A, B, Ç, ğ

Rakam yazıyoruz → 1, 2, 3

O zaman soru: Bilgisayar "A" harfini hangi sayıya çevirecek? "ç" harfi kaç olacak?


İşte buna karar veren kurala: Character Encoding denir.

ASCII de en eski encodinglerden biridir.
