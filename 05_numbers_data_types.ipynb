Print Fonksiyonu

print() fonksiyonu, programdaki bir değeri veya metni ekrana yazdırmak için kullanılır.

İnput Fonksiyonu

1- input() fonksiyonu, Python’da kullanıcıdan klavyeyle veri almak için kullanılır. 

2- input() fonksiyonu, program çalışırken kullanıcıdan veri girmesini ister ve girilen değeri string (str) olarak döndürür. 

3- Eğer kullanıcıdan int almak istiyorsan -> age = int(input("Yaşınızı girin: ")) 

4- Eğer float almak istiyorsan -> price = float(input("Fiyatı girin: ")) 

5- Küçük ama çok önemli hata örneği -> x = input("Bir sayı girin: ") print(x + 5) #Hata verir

Dairenin Alanını ve Çevresini Bulan Program

```python
pi = 3.1459
r = int(input("r'yi giriniz: "))
alan = pi * r * r
cevre = 2 * pi * r
print(alan)
print(cevre)
```
