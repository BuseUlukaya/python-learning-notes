STRINGLERDE DİLİMLEME VE BİRLEŞTİRME

String dilimleme, bir metnin (string’in) belirli bir parçasını almak demektir.

Python’da genel form:

string[start : stop : step]

start → Başlangıç indeksi (dahil)

stop → Bitiş indeksi (hariç)

step → Kaçar kaçar ilerleyeceği

Temel Dilimleme Örnekleri

1-Baştan belli bir yere kadar

```python
s = "Bilgisayar"
print(s[0:4])
```

2-Ortadan parça almak

```python
print(s[3:7])
```

3-Baştan sona kadar

```python
print(s[:10])
```

4-Belli bir yerden sona kadar

```python
print(s[5:])
```

5-Step Adım Kullanımı

```python
print(s[0:10:2]) #İkişer atlayarak
```

6-Ters Çevirme

```python
print(s[::-1])
```

7-Negatif İndeks İle Dilimleme

print(s[-5:])  #sondan 5 karakter

```python
print(s[-7:-3]) #sondan 7. karakterden sondan 3. karaktere kadar
```

STRING METODLARI

upper() -> Tüm harfleri BÜYÜK yapar.

```python
s = "merhaba"
s.upper()   # "MERHABA"
```

lower() -> Tüm harfleri küçük yapar.

```python
s = "MerHaBa"
s.lower()   # "merhaba"
```

strip() -> Baş ve sondaki boşlukları siler.

```python
s = "   merhaba   "
s.strip()   # "merhaba"
```

capitalize() -> Sadece ilk harfi büyük yapar.

```python
s = "merhaba dünya"
s.capitalize()   # "Merhaba dünya"
```

title() -> Her kelimenin ilk harfini büyük yapar.

```python
s = "merhaba dünya"
s.title()   # "Merhaba Dünya"
```

replace(eski, yeni) -> String içinde değiştirme yapar.

```python
s = "merhaba dünya"
s.replace("dünya", "python")   # "merhaba python"
```

STRING FORMATLAMA

```python
age = 21
name = "Buse Ulukaya"
print(f"My name is {name}, I am {age} years old.")
```

```python
price = 19.56789
print(f"Price:{price}")
```

Biz noktadan sonraki kısmı istediğimiz basamak sayısı kadar düzenleyebiliriz.

```python
print(f"price:{price:.2f}")
```

Noktadan sonra istedediğimiz basamak sayısını yukarıdakı gibi belirleyebiliriz.
