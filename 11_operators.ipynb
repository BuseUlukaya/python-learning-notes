**=	Üs al ve ata	x **= 3 → x = x ** 3

-=	Çıkar ve ata	x -= 3 → x = x - 3

//=	Tam böl ve ata	x //= 3 → x = x // 3

Bitwise (Bitsel) Atama Operatörleri

Bit kaydırma -> <<= -> x = x << y

MANTIKSAL OPERATÖRLER

1.And Operatörü

Her iki koşul da True ise → True

Biri bile False ise → False

```python
x = 7

x > 5 and x < 10   # True
x > 5 and x > 10   # False
```

2.Or Operatörü

En az bir koşul True ise → True


İkisi de False ise → False

```python
x = 3

x < 5 or x > 10   # True
x > 5 or x > 10   # False
```

3.Not Operatörü

Sonucu tersine çevirir

```python
x = 7

not(x > 5)        # False
not(x < 5)        # True
```
